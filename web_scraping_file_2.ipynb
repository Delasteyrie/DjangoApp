```python
#importer les packages
from newspaper import Article
import nltk
import pandas as pd
```

```python
#spécifier les articles qui m'interressent
urls = ["https://geoconfluences.ens-lyon.fr/glossaire/telecommunications-et-technologies-de-linformation-et-de-communication-tic#:~:text=Les%20t%C3%A9l%C3%A9communications%2C%20%C3%A9tymologiquement%20%C2%AB%20communication%20%C3%A0,celui%20des%20ondes%20%C3%A9lectro-magn%C3%A9tiques.",
       "https://geoconfluences.ens-lyon.fr/geoconfluences/informations-scientifiques/dossiers-thematiques/mobilites-flux-et-transports/corpus-documentaire/technologies-de-linformation-et-de-communication-quels-effets-sur-les-territoires",
       "http://geoconfluences.ens-lyon.fr/glossaire/telecommunications-et-technologies-de-linformation-et-de-communication-tic",
       "https://junior.universalis.fr/identification/",
       "https://hellofuture.orange.com/fr/mot-de-linnovation-coeur-de-reseau/",
        "https://geoconfluences.ens-lyon.fr/informations-scientifiques/dossiers-thematiques/mobilites-flux-et-transports/articles-scientifiques/transport-aerien-covid-euroairport-bale"]

def url_text(url):
  article=Article(url)
  article.download()
  article.parse()
  title=article.title
  auteur=article.authors
  text=article.text
  print(url)
  return title,auteur,text

texts =[url_text(url) for  url in urls]
```
